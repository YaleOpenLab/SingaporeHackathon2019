# seeing if can merge with current data:

```python
# import os
# files=os.listdir('htmlPages')
ground=pd.read_csv('ground.csv')
ground['TIMESTAMP'] = ground['TIMESTAMP'].map(lambda x: str(x)[:-6])
ground
```

```python
timestamp=pd.DataFrame(ground.TIMESTAMP)
timestamp
```

```python
import re
d={'Humidity':[]}
cache=0
cacheId=0
for i in range(len(timestamp)):
    s=timestamp.loc[i][0]
    patternStr=r'-(\d+)'
    L=re.findall(patternStr,s)
    fileName='-'.join([re.sub(r'0(\d)',r'\1',x) for x in L]) # using regex groups: converting type 01 -> 1, 10 -> 10
    table= pd.read_html('htmlPages/{}.html'.format(fileName))[1]

    humidity=table[table.columns[6]]
    humidity =humidity.drop(0);humidity =humidity.drop(len(humidity))
    humidity=humidity[humidity.notnull()]
    humidity=humidity.map(lambda x: int(x[:-1]))
    d['Humidity'].append(humidity.mean())

humidityDf=pd.DataFrame(d)
humidityDf
```
