```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("/Users/yuesongyang/Desktop/Hack4Climate/GCAM.xlsx", sheet_name="GCAM_Timeseries data")
```

```python
countries=["China", "India","Brazil","Canada","EU-12","United States","Japan"]
```

```python
df=df[df["Region"].isin(countries)]
```

```python
df.describe()
```

```python
nrow=3192
```

```python
def diff(x):
    return x[1]-x[0]
```

```python
df.head()
```

```python
row=df.iloc[0,5:7].tolist()
row
```

```python
s = np.random.normal(0, 1, nrow)
```

```python
#2005-2010
new_col=[]
for i in range (0,nrow):
    row=df.iloc[i,5:7].tolist()
    res=diff(row)
    step_size=res/5
    new_col.append(step_size)

for j in range(1,5):
    lst=[]
    for i in range(0,nrow):
        lst.append(j * float(new_col[i]) + df.iloc[:,5:6].values.tolist()[i][0] + float(s[i]))
    year=j+2005
    df[str(year)]=lst
```
