```python
df3["overall score"] = (df3["Climate Risk Score"] + df3["Vulnerability score"] + df3["Readiness score"])/3
```

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y')
data = pd.read_csv('/Users/ZhengShanshan/Desktop/CO2_emission(1850-2014).csv', parse_dates=['Year'], index_col='Year',date_parser=dateparse)
```

```python
USA = ['United States of America']
data.Country.isin(USA)
USA_data = data[data.Country.isin(USA)]
ts_USA = USA_data['Total CO2 Emissions Excluding Land-Use Change and Forestry (MtCO2)'].tolist()
for i in range(36):
    model = ARIMA(ts_USA, order=(2, 1, 0))  
    model_fit = model.fit()
    output = model_fit.forecast()
    yhat = output[0][0]
    ts_USA.append(yhat)
ts_USA = ts_USA[-37: -1] # the predicted value for USA
    
```
